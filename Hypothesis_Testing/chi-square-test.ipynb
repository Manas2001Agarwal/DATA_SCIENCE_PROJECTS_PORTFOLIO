```python
dataset_table = pd.crosstab(data['Level_of_Stress'],data['Blood_Pressure_Abnormality'])
dataset_table
```

> H0 :- Level_of_Strees and Blood_Pressure_Abnormality are related to each other


> H1 :- Level_of_Strees and Blood_Pressure_Abnormality are not related to each other

```python
import scipy.stats as stats
Expected_val=stats.chi2_contingency(dataset_table)
Expected_val = Expected_val[3]
```

```python
print(Expected_val)
```

```python
no_of_rows=len(dataset_table.iloc[:,0])
print("no_of_rows :- ",no_of_rows)

no_of_columns=len(dataset_table.iloc[0,:])
print("no_of_columns :- ",no_of_columns)

ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05
```

```python
#Observed Values
Observed_Values = dataset_table.values 
print("Observed Values :-\n",Observed_Values)
```

```python
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_val)])
chi_square_statistic=chi_square[0]+chi_square[1]
print(chi_square_statistic)
```

```python
alpha = 0.05 # Significance value
```

```python
from scipy.stats import chi2
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
```

```python
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
print('p-value:',p_value)
```

```python
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables, ie: Level_of_Strees and Blood_Pressure_Abnormality are related to each other")
else:
    print("Retain H0,There is no relationship between 2 categorical variables, ie: Level_of_Strees and Blood_Pressure_Abnormality are not related to each other")
print(" ")
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables, ie: Level_of_Strees and Blood_Pressure_Abnormality are related to each other")
else:
    print("Retain H0,There is no relationship between 2 categorical variables, ie: Level_of_Strees and Blood_Pressure_Abnormality are not related to each other")
```

This Statistical test declares that there is no relationship between level of stress and blood pressure levels. But we must remember that these test are based only on limited available data and may be wrong
