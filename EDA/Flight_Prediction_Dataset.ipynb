**Now we will take or derive some features for example from date we will extract date, month and year**

```python
final_df['Date_of_Journey'].str.split('/').str[0]
```

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0].astype(int)
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1].astype(int)
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis= 1,inplace = True)
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
final_df
```

```python
final_df['Hour'] = final_df['Arrival_Time'].apply(lambda x:x.split(':')[0])
final_df['Minutes'] = final_df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

```python
final_df.drop('Arrival_Time',inplace = True,axis=1)
```

```python
final_df
```

```python
final_df['Hour'] = final_df['Hour'].astype(str)
final_df['Minutes'] = final_df['Minutes'].astype(str)
```

```python
final_df['Dept_Hour'] = final_df['Dep_Time'].apply(lambda x:x.split(':')[0]).astype(str)
final_df['Dept_Minutes'] = final_df['Dep_Time'].apply(lambda x:x.split(':')[1]).astype(str)

final_df.drop('Dep_Time',inplace = True,axis=1)
```

```python
final_df
```

```python
final_df['Route'].value_counts()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
final_df
```

```python
final_df.drop('Route',axis = 1,inplace = True)
```

```python
final_df[final_df['Duration']== '5m' ]
```

```python
final_df.drop(index = 6474, inplace = True)
```

```python
hour = final_df['Duration'].str.split('h').str[0].astype(int)
hour = hour*60
minutes = final_df['Duration'].str.split('h').str[1].astype(str)
minutes = minutes.str.split(' ').str[1]
```

```python
minutes = minutes.str.split('m').str[0]
```

```python
hour
```

```python
hour_m = hour
```

```python
for i in minutes.index:
  if minutes[i] is not np.nan:
    minutes[i] = int(minutes[i])
    hour_m[i] = hour[i]+minutes[i]
  else:
    hour_m[i] = hour[i]
```

```python
final_df['Duration_in_min'] = hour_m
```

```python
hour_m.shape
```

```python
final_df.drop('Duration',axis = 1,inplace = True)
```
