Alternative Approach for missing values

```python
dataset  = df
```

```python
cat_3 = df.groupby(['Age'])['Product_Category_2'].median()
cat_3 = cat_3.to_dict()
cat_3
```

```python
import math

for i in dataset.index:
  
  if math.isnan(dataset.loc[i]['Product_Category_2']):

      value = cat_3[dataset.loc[i]['Age']]
      dataset.loc[i,'Product_Category_2'] = value
    
```
